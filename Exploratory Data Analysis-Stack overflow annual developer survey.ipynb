```python
#Analysing companies with developers having the most pay
# Removing rows with missing values
stack_df.dropna(subset=['OrgSize', 'ConvertedCompYearly'], inplace=True)
```

```python
# Group by 'OrgSize and calculate median salary for different groups
salary_org_size = stack_df.groupby('OrgSize')['ConvertedCompYearly'].median()
```

```python
# Sort the results to identify which organization size group has the highest median salary
highest_paid_orgsize = salary_org_size.sort_values(ascending=False)
```

```python
# Display the result
print("Organization Size with Highest Median Salary:")
print(highest_paid_orgsize)
```

```python
#Visualisation for the companies with developers with the most pay
plt.figure(figsize=(14, 8))
sns.barplot(x = salary_org_size.values, y = salary_org_size.index, orient='h', edgecolor = 'r')
plt.title('Median Yearly Compensation by Organization Size')
plt.xlabel('Median Converted Compensation (Yearly)')
plt.ylabel('Organization Size')
plt.show()
```

```python
# Analysing and visualising the remotework column and how it matters to employees

remote_work_count = stack_df['RemoteWork'].value_counts()
remote_work_count

```

```python
# Plotting a pie chart
colors = ['Green', 'Salmon']
plt.figure(figsize=(8, 6))
plt.pie(remote_work_count, labels=remote_work_count.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Remote Work Preference')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#Line plot to show how average yearly compensation varies based on years of professional eperience. 

stack_df['YearsCodePro'] = pd.to_numeric(stack_df['YearsCodePro'], errors='coerce')

stack_df = stack_df.dropna(subset=['ConvertedCompYearly'])

average_salary_by_years = stack_df.groupby('YearsCodePro')['ConvertedCompYearly'].mean().reset_index()

plt.figure(figsize=(12, 8))
sns.lineplot(x='YearsCodePro', y='ConvertedCompYearly', data=average_salary_by_years)
plt.title('Average Salary vs Years of Coding Professionally')
plt.xlabel('Years of Coding Professionally')
plt.ylabel('Average Converted Compensation (Yearly)')
plt.show()
```

```python
#Analysing the most popular methods in learning coding
# Fill missing values with empty strings and concatenate columns
stack_df['CombinedLearnCode'] = (
    stack_df['LearnCode'].fillna('') + ';' +
    stack_df['LearnCodeOnline'].fillna('') + ';' +
    stack_df['LearnCodeCoursesCert'].fillna('')
)
```

```python
# Splitting the combined string into a list of methods and strip whitespace
stack_df['CombinedLearnCode'] = stack_df['CombinedLearnCode'].apply(lambda x: [method.strip() 
for method in x.split(';')])
```

```python
#creating a list Aggr_learning_methods by iterating over each element.
Aggr_learning_methods = [method for sublist in stack_df['CombinedLearnCode'] 
for method in sublist]
```

```python
# creating a Pandas Series named learning_methods_counts
learning_methods_counts = pd.Series(Aggr_learning_methods).value_counts()
```

```python
#plotting a barplot
plt.figure(figsize=(14, 8))
sns.barplot(x=learning_methods_counts.values, y=learning_methods_counts.index, orient='h', edgecolor='skyblue')
plt.title('Popular Methods of Learning to Code')
plt.xlabel('Number of Respondents')
plt.ylabel('Learning Methods')
plt.show()
```

```python
#Analysis for Employed developers by Educational level
#Removing null values from the selected columns:'Edlevel, and Employment'

stack_df_filtered = stack_df[['EdLevel', 'Employment']].dropna(subset=['EdLevel', 'Employment'])
```

```python
# creating a new column named 'IsEmployed' in the DataFrame stack_df_filtered.
stack_df_filtered['IsEmployed'] = stack_df_filtered['Employment'].apply(lambda x: 'Employed' in x)
```

```python
#calculating the percentage of employed individuals within different educational levels 
employment_by_education = stack_df_filtered.groupby('EdLevel')['IsEmployed'].mean() * 100
```

```python
#sorting the employment_by_education Series in descending order based on the values
employment_by_education = employment_by_education.sort_values(ascending=False)
```

```python
#plotting for visualisation
plt.figure(figsize=(12, 8))
sns.barplot(x=employment_by_education, y=employment_by_education.index, orient='h')
plt.title('Percentage of Employed Developers by Education Level')
plt.xlabel('Percentage Employed')
plt.ylabel('Education Level')
plt.show()
```

```python
#Analysing years of experience needed to get a remote developer job.
# Filter data for remote workers
remote_workers = stack_df[(stack_df['RemoteWork'] == 'Yes') | (stack_df['Employment'].str.contains('Remote'))]
```

```python
#clean the YearsCodePro
remote_workers['YearsCodePro'] = pd.to_numeric(remote_workers['YearsCodePro'], errors='coerce')
```

```python
# Droping rows with missing values in relevant columns
remote_workers = remote_workers.dropna(subset=['YearsCodePro'])
```

```python
#  calculating the Descriptive statistics
print(remote_workers['YearsCodePro'].describe())
```

```python
#checking for correlation between years of professional coding and organisation
correlation = remote_workers['YearsCodePro'].corr(remote_workers['OrgSize'])
print("Correlation between Years of Professional Coding Experience and Organisation Size:", correlation)
```
